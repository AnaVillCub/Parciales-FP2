## Esquema Extendido: Listas Encadenadas

### I. Conceptos Fundamentales
- Definición: Estructura formada por un conjunto de objetos (nodos) que se referencian unos a otros.
- Puntos de Control:
    - Objeto Inicial: Punto de entrada a la estructura (ej. first, cima, raíz). Si se pierde esta referencia, toda la estructura queda inaccesible.
    - Objetos Terminales: Nodos finales que no referencian a otros objetos (su puntero es None o null).
- Naturaleza en Python: Se basa en referencias (direcciones de memoria). Los atributos de un objeto son, a su vez, referencias a otros objetos.

### II. Estructura de Datos (Implementación)
- __Clase Nodo (Node)__: Generalmente definida como una clase anidada dentro del contenedor
    - .value: El dato almacenado.
    - next_node: Referencia al siguiente nodo.
- __Clase Contenedora (LinkedList)__:
    - Atributo _first: Referencia al primer elemento.
    - Atributo _last (opcional): Optimiza la inserción al final.
    - Atributo _len: Almacena la longitud para evitar recorrer la lista cada vez que se consulte.

### III. Tipos de Estructuras según el PDF
1. __Estructuras Recursivas__: Un objeto tiene una referencia a otro objeto de su misma clase (ej. una Persona que referencia a su padre que también es Persona).
2. __Múltiples Secuencias__: Un mismo conjunto de objetos puede tener varios atributos de encadenamiento para permitir diferentes órdenes (ej. un puntero para orden alfabético y otro para orden de longitud).

### IV. Funcionalidades Avanzadas en Python
- Iteración: Para que una LinkedList sea iterable, se debe implementar el método __iter__() como una función generadora usando la palabra clave yield.

```python
def __iter__(self):
    current = self._first
    while current != None:
        yield current.value
        current = current.next_node
```

## OPERACIONES PRINCIPALES

### 1. Inserción (Agregar elementos)
La inserción varía enormemente en eficiencia dependiendo de dónde colocas el dato.
- __Al Principio (First)__: Solo creas un nodo nuevo y haces que su referencia next_node apunte al que antes era el primero.
- __Al Final (Append)__: * Si solo tienes referencia al primero, debes recorrer toda la lista hasta el final.
    - Si la clase tiene un atributo _last, se vuelve inmediata.
- __En el Medio__: Requiere buscar el nodo justo anterior a la posición deseada. El nuevo nodo se convierte en el "sucesor" del anterior y en el "predecesor" del que seguía.

```python
def insert_as_first(self, value):
    # El nuevo nodo apunta al actual primero
    self._first = self.Node(value, self._first) 
    self._len += 1 # Actualizamos el contador
```


### 2. Eliminación (Borrar elementos)
Para borrar, el objetivo es "saltarse" el nodo que ya no queremos, redirigiendo los punteros.
- __Borrar el Primero__: Simplemente haces que self._first apunte al segundo nodo (self._first.next_node).
- __Borrar el Último__: Necesitas encontrar al penúltimo para que pase a ser el nuevo final y apunte a None.
- __Por Posición__: Se recorre la lista hasta el nodo anterior al que se quiere eliminar. Se cambia su referencia para que "salte" el nodo objetivo y apunte directamente al siguiente.

    - Ejemplo de lógica (Borrado por posición):
        - Si quieres borrar el nodo en el índice 2, haces que el nodo 1 apunte directamente al nodo 3.


### 3. Búsqueda y Acceso
A diferencia de los arreglos (arrays), aquí no hay "atajos".
- __Acceso Secuencial__: Para encontrar un dato o saber la longitud (si no guardas el atributo _len), debes empezar en _first y usar un bucle while para saltar de nodo en nodo hasta que el puntero sea None.
- __Desventaja__: Es más lento que un array porque no puedes calcular la posición en memoria directamente; tienes que pasar por todos los predecesores.

### 4. Iteración (Recorrido)
Una forma elegante de recorrer la lista en Python es usando el método mágico __iter__.
- __Función Generadora__: Se usa yield para devolver los valores uno a uno sin destruir la lista ni crear una copia pesada en memoria.

__Ejemplo de esquema de iteración__:
1. Empiezas en current = self._first.
2. Mientras current no sea None:
    - Entregas el valor (yield current.value).
    - Avanzas al siguiente (current = current.next_node).

### IV. Operaciones Principales y Ejemplos
1. __Inserción por Delante (insert_as_first)__
- Lógica: Se crea un nuevo nodo cuya referencia next_node apunta al que era el primer nodo hasta ese momento.
Código clave: self._first = self.Node(value, self._first).

2. __Inserción por Detrás (append)__
- Caso A (Sin puntero al final): Se debe recorrer toda la lista con un bucle while current.next_node != None hasta localizar el último nodo. 
- Caso B (Con puntero _last): Se accede directamente al último nodo, se conecta el nuevo y se actualiza la referencia _last. 

3. __Borrado del Primer Elemento (delete_first)__
- Lógica: La referencia _first de la lista pasa a apuntar al segundo nodo (self._first.next_node).
- Importante: Se debe actualizar el contador _len y, si la lista queda vacía, poner _last en None.

4. __Borrado del Último Elemento (delete_last)__
- Lógica: Es costosa incluso con puntero _last, porque necesitas encontrar el penúltimo elemento para convertirlo en el nuevo "objeto terminal".
- Proceso: Se usa un puntero prev que sigue a current mientras se recorre la lista hasta que current llega a _last.

5. __Búsqueda e Iteración__
- Acceso: No es directo como en un array. Para llegar al elemento $i$, hay que pasar por los $i-1$ anteriores.
- Iteración en Python: Se implementa el método __ iter __ como una función generadora usando yield. Esto permite usar la lista en un bucle for item in lista.

## Gestión de Cola en un Supermercado 
Se pide finalizar el desarrollo de la clase CheckoutQueue del módulo supermarket. Cada objeto de dicha clase representa la cola de clientes de una caja en un supermercado. Se proporciona la clase Client completada; y la clase CheckoutQueue por completar que ya incluye un constructor básico y una clase interna Node a usar para los nodos de la lista encadenada conteniendo los clientes de la cola. 

Cada objeto de la clase Client representa a un cliente empleando dos atributos públicos: name para el nombre y total_cost para el total a pagar por su compra.

Para completar la clase CheckoutQueue deberás desarrollar los siguientes métodos y propiedades (pudiendo para ello añadir o modificar sus actuales atributos o métodos si fuera necesario):

✅ Métodos y propiedades a desarrollar:
- Método enqueue: Representa la llegada de un cliente (recibido por parámetro) a la cola, añadiéndolo por el final.
- Método dequeue: Representa el que un cliente ha pasado por caja. Para ello, quita el cliente que estaba al principio de la cola y lo devuelve. Además, ha de actualizar apropiadamente lo ingresado en esa caja desde que se abrió su cola. Devuelve None si no había cliente en la cola.
- Propiedad daily_sales: Esta propiedad de solo lectura nos devuelve lo ingresado en esa caja desde que se abrió la cola.
- Método split: Este método está pensado para que los clientes de la cola pasen a ser atendidos en dos cajas. Para ello crea una nueva cola y la devuelve tras repartir los clientes, de forma que:
    - La cola original conservará los clientes de posiciones pares (0, 2, 4, ...),
    - La nueva cola contendrá los clientes de posiciones impares (1, 3, 5, ...),
    - Manteniendo en ambas el orden relativo original de sus clientes.

```python
from containerbase import ContainerBase # type: ignore

class Client:

    def __init__(self, name, total_cost):
        self.name = name
        self.total_cost = total_cost

    def __str__(self):
        return f"{self.name}: {self.total_cost}"


class CheckoutQueue(ContainerBase):
    class Node:
        def __init__(self, value, next=None):
            self.value = value  # El valor en el nodo.
            self.next = next    # El enlace al siguiente nodo.

    def __init__(self):
        self.first = None   # El enlace al primer nodo de la lista.
        self.last = None   # El enlace al último nodo de la lista.
        self.__daily_sales = 0
    # Método enqueue: Representa la llegada de un cliente (recibido por parámetro) a la cola, añadiéndolo por el final.
    def enqueue(self, cliente):
        nuevo_cliente = self.Node(cliente)
        if self.first is None: # Esta vacia la lista
            self.first = nuevo_cliente
        else:
            self.last.next = nuevo_cliente # La lista está llena, ponemos al siguiente del final
        self.last  = nuevo_cliente # Reafirmamos q el ultimo es el nuevo cliente

    # Método dequeue: Representa el que un cliente ha pasado por caja. Para ello, quita el cliente que estaba al principio de la cola y lo devuelve. 
    # Además, ha de actualizar apropiadamente lo ingresado en esa caja desde que se abrió su cola. Devuelve None si no había cliente en la cola.
    def dequeue(self):
        if self.first is None:
            return None
        cliente = self.first.value # Cogemos el primero nodo, y extraemos el valor (name, total_cost)
        self.first = self.first.next # Vamos al siguiente nodo
        if self.first is None:
            self.last = None
        self.__daily_sales += cliente.total_cost # Cliente esta definido anteriormente, y ahora accedemos al total_cost
        return cliente
        
    # Propiedad daily_sales: Esta propiedad de solo lectura nos devuelve lo ingresado en esa caja desde que se abrió la cola.
    @property
    def daily_sales(self):
        return self.__daily_sales
    
    # Método split: Este método está pensado para que los clientes de la cola pasen a ser atendidos en dos cajas. Para ello crea una nueva cola y la devuelve tras repartir los clientes, de forma que:
    #     - La cola original conservará los clientes de posiciones pares (0, 2, 4, ...),
    #     - La nueva cola contendrá los clientes de posiciones impares (1, 3, 5, ...),
    #     - Manteniendo en ambas el orden relativo original de sus clientes.
    def split(self):
        nueva_cola = CheckoutQueue()
        if self.first is None:
            return None
        actual = self.first
        temporales = [] # Creamos una lista para añadir los clientes

        while actual is not None:
            temporales.append(actual.value)
            actual = actual.next # Avanzamos
        # Reiniciamos la lista
        self.first = None
        self.last = None

        for i, cliente in enumerate(temporales):
            if i%2 == 0:
                self.enqueue(cliente)  # En la lista self, mantenemos pares
            else:
                nueva_cola.enqueue(cliente)
        return nueva_cola
```

## Gestión de Grupos de Interés
Se pide completar el desarrollo de la clase InterestGroup en el módulo networking. Cada objeto de la clase InterestGroup representa un grupo de personas con intereses comunes. Para almacenar los miembros del grupo se usará una lista encadenada. Se proporciona la clase interna Node que define las características de los nodos de la lista. Los miembros del grupo están representados por sus nombres (tipo str).

La clase InterestGroup hereda de la clase ContainerBase, que define el método __str__, permitiendo mostrar la lista de miembros de forma sencilla.

Tu tarea es completar el funcionamiento de la clase InterestGroup añadiendo atributos y modificando métodos que crea necesarios con el fin de desarrollar los siguientes métodos y propiedades dentro de la clase InterestGroup:

✅ Métodos y propiedades a desarrollar:
Método is_member: Devuelve True si el nombre pasado por parámetro está en el grupo, False en caso contrario.

Método add_member: Añade un nuevo miembro (nombre pasado por parámetro) al grupo si no estaba previamente. Devuelve True si lo añade y False si no.

Propiedad size: Devuelve el número total de miembros del grupo.

Método union: Devuelve un nuevo grupo con la unión de miembros del grupo actual y el grupo recibido por parámetro (sin duplicados).

Método remove_member: Se le pasa un nombre por parámetro y lo elimina si está en el grupo. Devuelve True si lo elimina y False si no.

```python
from containerbase import ContainerBase  # type: ignore

class InterestGroup(ContainerBase):

    class Node:
        def __init__(self, value, next=None):
            self.value = value
            self.next = next

    def __init__(self):
        self.first = None
        self.__size = 0

    # Propiedad size: Devuelve el número total de miembros del grupo.
    @property
    def size(self):
        cont = 0
        actual = self.first
        while actual:
            cont += 1
            actual = actual.next
        return cont

    # Método is_member: Devuelve True si el nombre pasado por parámetro está en el grupo, False en caso contrario.
    def is_member(self, miembro):
        actual = self.first
        while actual is not None:
            if actual.value == miembro:
                return True
            actual = actual.next
        return False

    # Método add_member: Añade un nuevo miembro (nombre pasado por parámetro) al grupo si no estaba previamente. Devuelve True si lo añade y False si no.
    def add_member(self, miembro):
        if self.is_member(miembro):
            return False
        self.first = self.Node(miembro, self.first)
        self.__size += 1
        return True

    # Método remove_member: Se le pasa un nombre por parámetro y lo elimina si está en el grupo. Devuelve True si lo elimina y False si no.
    def remove_member(self, miembro):
        actual = self.first
        anterior = None
        while actual is not None:
            if actual.value == miembro:
                if anterior is None:
                    self.first = actual.next  # borra el primer nodo
                else:
                    anterior.next = actual.next  # borra nodo intermedio/final
                self.__size -= 1
                return True
            anterior = actual
            actual = actual.next
        return False

    # Método union: Devuelve un nuevo grupo con la unión de miembros del grupo actual y el grupo recibido por parámetro (sin duplicados).
    def union(self, otro):
        nuevo_grupo = InterestGroup()
        actual = self.first
        while actual is not None:
            nuevo_grupo.add_member(actual.value)
            actual = actual.next
            
        actual_otro = otro.first
        while actual_otro is not None:
            nuevo_grupo.add_member(actual_otro.value)
            actual_otro = actual_otro.next
        return nuevo_grupo
```

```python
from containerbase import ContainerBase  # type: ignore

class Task:

    def __init__(self, name, duration):
        self.name = name
        self.duration = duration

    def __str__(self):
        return f"{self.name}: {self.duration} min"


class TaskList(ContainerBase):

    class Node:
        def __init__(self, value, next=None):
            self.value = value
            self.next = next

    def __init__(self):
        self.first = None
        self.last = None
        self._size = 0
        self._time_spent = 0
    
    # Método add_task: Añade una nueva tarea pasada por parámetro al final de la lista.
    def add_task(self, task):
        new_node = self.Node(task)
        if self.first is None:
            self.first = new_node
            self.last = new_node
        else:
            self.last.next = new_node # type: ignore
            self.last = new_node
        self._size += 1
    
    # Método get_task: Devuelve la tarea que se encuentra en la posición pasada por parámetro 
    # (comenzando desde 0). Si la posición no es válida, devuelve None.
    def get_task(self, position):
        if position < 0 or position >= self._size: # Verificamos los límites
            return None
        current = self.first
        for _ in range(position):
            if current is None: # Si la lista está mal construida
                return None
            current = current.next
        return current.value if current else None # Asegura que current no sea None
    
    # Método task_completed: Elimina y devuelve la tarea en la posición dada como parámetro, si existe. 
    # Esta acción debe considerar la propiedad time_expent. Devuelve None si la posición es incorrecta.
    def task_completed(self, position):
        if position < 0 or position >= self._size:
            return None
        
        if position == 0:
            if self.first is None: # Si la lista està vacía
                return None
            completed_task = self.first.value
            self._time_spent += completed_task.duration
            self.first = self.first.next
            if self.first is None:
                self.last = None
        else:
            current = self.first
            for _ in range(position - 1):
                if current is None or current.next is None: #  Si la posición no existe
                    return None
                current = current.next
                if current.next is None:  # Si el siguiente nodo no existe
                    return None
            completed_task = current.next.value # type: ignore
            self._time_spent += completed_task.duration
            current.next = current.next.next # type: ignore
            if current.next is None: # type: ignore
                self.last = current
        
        self._size -= 1
        return completed_task
    
    # Propiedad de solo lectura time_left: Devuelve el tiempo total (en minutos) de todas las tareas pendientes en la lista.
    @property
    def time_left(self):
        total = 0
        current = self.first
        while current is not None:
            total += current.value.duration
            current = current.next
        return total
    
    # Propiedad de solo lectura time_spent: Devuelve el tiempo total que se ha dedicado a las tareas completadas.
    @property
    def time_spent(self):
        return self._time_spent
```

### Sistema de Gestión de Playlist (Reproductor de Música)
Se pide desarrollar la clase Playlist para gestionar una lista de canciones en espera. Cada canción está representada por un objeto de la clase Song, la cual dispone de dos atributos: title (nombre de la canción) y artist (nombre del cantante o banda).Para facilitar la visualización, la clase Playlist hereda de BaseContainer. Internamente, las canciones deben almacenarse en una lista enlazada. Se proporciona una clase interna Node para los elementos de esta lista.

✅ Métodos y propiedades a desarrollar:
- Método add_to_queue(self, song):
    - Añade una nueva canción al final de la playlist. Se recomienda usar un atributo para el último nodo.
- Método play_next(self):
    - Representa que la primera canción de la lista ha terminado de sonar y debe ser eliminada. El método debe devolver el objeto Song eliminado. Si la playlist está vacía, devuelve None.
- Método search_by_artist(self, artist_name):
    - Debe devolver una nueva Playlist que contenga únicamente las canciones del artista indicado, manteniendo el orden original. Las canciones no deben eliminarse de la playlist original.
- Método remove_by_title(self, title):
    - Busca en la playlist una canción con el título indicado y la elimina si existe. Debe devolver True si se encontró y eliminó con éxito, o False en caso contrario.
- Propiedad de solo lectura is_empty
    - Devuelve True si no hay canciones en la playlist y False si contiene al menos una.
    
🧠 Restricciones y Recomendaciones:
- Prohibido el uso de listas nativas de Python ([] o list()) para almacenar las canciones.
- Recuerda que si eliminas la canción que está en la posición first, debes actualizar dicha referencia.
- Si eliminas la última canción de la lista, asegúrate de actualizar correctamente la referencia al último nodo.
- Ten especial cuidado al borrar: necesitas conocer el nodo anterior al que quieres eliminar para "saltarlo" correctamente.

```python
from containerbase import ContainerBase # type: ignore

class Song:
    def __init__(self, title, artist):
        self.title = title
        self.artist = artist

    def __str__(self):
        return f"{self.title} - {self.artist}"

class Playlist(ContainerBase):
    class Node:
        def __init__(self, value, next=None):
            self.value = value  # Aquí guardamos el objeto Song
            self.next = next    # Referencia al siguiente nodo 

    def __init__(self):
        self.first = None   # Punto de entrada a la estructura 
        self.last = None    # Referencia al último para inserción eficiente O(1)
        # Contador para la propiedad size [cite: 187]
        self._size = 0      

    # Propiedad de solo lectura is_empty: Devuelve True si no hay canciones en la playlist y False si contiene al menos una.
    @property
    def is_empty(self):
        """Devuelve True si la lista no tiene nodos."""
        return self.first is None

    # Método add_to_queue(self, song): Añade una nueva canción al final de la playlist. Se recomienda usar un atributo para el último nodo.
    def add_to_queue(self, song):
        """
        Añade la canción al final. 
        Recuerda: si la lista está vacía, first y last apuntan al nuevo.
        """
        new_song = self.Node(song)
        if self.first is None:
            self.first = new_song
        else:
            self.last.next = new_song
        self.last = new_song
        self._size += 1

    # Método play_next(self): Representa que la primera canción de la lista ha terminado de sonar y debe ser eliminada. El método debe devolver el objeto Song eliminado. Si la playlist está vacía, devuelve None.
    def play_next(self):
        """
        Elimina el primer nodo y devuelve la canción.
        No olvides actualizar el puntero 'last' si la lista se vacía.
        """
        if self.is_empty:  # Paso 1: Verificar si está vacía
            return None
        song = self.first.value  # Paso 2: Guardar la canción para devolverla
        self.first = self.first.next  # El nuevo 'first' es el que era el segundo
        self._size -= 1
        # Si la lista se vació completamente, actualizamos 'last'
        if self.first is None:
            self.last = None
        return song

    # Método search_by_artist(self, artist_name): Debe devolver una nueva Playlist que contenga 
    # únicamente las canciones del artista indicado, manteniendo el orden original. Las canciones no deben eliminarse de la playlist original.
    def search_by_artist(self, artist_name):
        new_Playlist = Playlist()
        actual = self.first
        while actual is not None:
            if (actual.value.artist == artist_name):
                new_Playlist.add_to_queue(actual.value)
            actual = actual.next
        return new_Playlist

    # Método remove_by_title(self, title): Busca en la playlist una canción con el título indicado y la elimina si existe. 
    # Debe devolver True si se encontró y eliminó con éxito, o False en caso contrario.
    def remove_by_title(self, title):
        if self.first is None: return False
        actual = self.first
        anterior = None
        while actual is not None:
            if actual.value.title == title:  # 1. Comparamos el título del objeto Song
                if anterior is None:  # 2. Si es el primer nodo
                    self.first = actual.next
                else:
                    anterior.next = actual.next  # 3. "Saltamos" el nodo actual conectando anterior con el siguiente
                if actual == self.last:  # 4. Si el nodo borrado era el último, actualizamos last
                    self.last = anterior
                self._size -= 1
                return True # Éxito: encontrado y eliminado
            anterior = actual  # 5. Si no coincide, avanzamos ambos punteros
            actual = actual.next
        return False # No se encontró después de recorrer toda la lista
```

### Sistema de Gestión de Inventario (Almacén)
Se pide completar la clase Inventory que gestiona los productos de un almacén. Cada producto está representado por la clase Product, que tiene tres atributos: id (código numérico), name (nombre) y stock (cantidad disponible).La clase Inventory debe usar internamente una lista enlazada de objetos Node.

✅ Métodos y propiedades a desarrollar:

- Propiedad total_inventory_value
    - Imagina que cada unidad de producto vale 10€. Devuelve el valor total del inventario sumando el stock de todos los productos y multiplicándolo por 10. 

```python
from containerbase import ContainerBase # type: ignore

class Product:
    def __init__(self, id, name, stock):
        self.id = id
        self.name = name
        self.stock = stock

class Inventory(ContainerBase):
    class Node:
        def __init__(self, value, next=None):
            self.value = value
            self.next = next

    def __init__(self):
        self.first = None # Punto de entrada [cite: 28, 31, 184]
        self._size = 0    # Contador de nodos [cite: 187]

    def add_new_product(self, product):
        """Añade al principio (ya lo sabes hacer, úsalo para probar)."""
        self.first = self.Node(product, self.first)
        self._size += 1

    # Método restock_product(self, id, quantity) Busca el producto con el id indicado. 
    # Si lo encuentra, aumenta su stock en la cantidad recibida y devuelve True. Si no existe, devuelve False.
    def restock_product(self, id, quantity):
        actual = self.first

        while actual is not None:
            if actual.value.id == id:
                actual.value.stock += quantity
                return True
            actual = actual.next
        return False

    # Método remove_low_stock(self, threshold): Elimina de la lista todos los productos cuyo stock sea estrictamente menor que el threshold (umbral) pasado por parámetro. 
    # Debe devolver el número total de productos eliminados.
    def remove_low_stock(self, threshold):
        actual = self.first
        contador = 0  # Cuenta borrados elementos
        anterior = None

        while actual is not None:
            if actual.value.stock < threshold:
                if anterior is None:  # CASO A: Es el primer nodo el que hay que borrar
                    self.first = actual.next
                else:
                    anterior.next = actual.next  # CASO B: Es un nodo intermedio o el último
                # Si el nodo que borramos era el último, actualizamos la referencia
                # (aunque este ejercicio no pedía puntero 'last', es buena práctica)
                self._size -= 1
                contador += 1
                actual = actual.next
            else:
                # Si NO eliminamos, avanzamos ambos punteros normalmente
                anterior = actual
                actual = actual.next
        return contador

    # Método get_most_stocked(self): Recorre la lista y devuelve el objeto Product que tenga la mayor cantidad de stock. 
    # Si la lista está vacía o hay empate, devuelve el primero que encuentre o None. 
    def get_most_stocked(self):
        if self.first is None:
            return None
        actual = self.first
        max_product = actual.value
        while actual is not None:
            if actual.value.stock > max_product.stock:
                max_product = actual.value
            actual = actual.next
        return max_product
    
    # Propiedad total_inventory_value: Imagina que cada unidad de producto vale 10€. Devuelve el valor total del inventario sumando el stock de todos los productos y multiplicándolo por 10. 
    @property
    def total_inventory_value(self):
        total = 0
        actual = self.first
        while actual is not None:
            total += actual.value.stock
            actual = actual.next
        return total*10
```

### Gestión de Turnos Bancarios (Prioridad)
Se pide completar la clase BankQueue. En este banco, los clientes no solo esperan su turno, sino que tienen un nivel de prioridad (1 para clientes VIP, 2 para clientes normales). La lista debe estar siempre ordenada: los VIP van delante de los normales.

✅ Métodos y propiedades a desarrollar:
- Método add_client(self, name, priority)
    - Añade un cliente a la lista.
    1. Si es VIP (prioridad 1), debe insertarse antes de cualquier cliente normal (prioridad 2).
    2. Si es Normal (prioridad 2), se añade al final de la lista.
    - Nota: Dentro del mismo nivel de prioridad, se respeta el orden de llegada.
- Método attend_client(self)
    - Elimina y devuelve el nombre del primer cliente en la lista (el que tiene más prioridad y lleva más tiempo esperando). 
    - Devuelve None si la cola está vacía.
- Método promote_to_vip(self, name)
    - Busca a un cliente por su nombre. Si es de prioridad 2, lo "asciende": lo elimina de su posición actual y lo reinserta al final de la sección VIP (justo antes del primer cliente normal).
- Propiedad vip_count
    - Devuelve cuántos clientes con prioridad 1 hay actualmente en la cola.

```python
class BankQueue(ContainerBase):
    class Node:
        def __init__(self, name, priority, next=None):
            self.name = name
            self.priority = priority # 1: VIP, 2: Normal
            self.next = next

    def __init__(self):
        self.first = None
        self._size = 0

    @property
    def size(self):
        return self._size

    # add_client: Añade un cliente a la lista. Si es VIP (prioridad 1), se inserta antes de los clientes normales; si es Normal (prioridad 2), se añade al final.
    # Dentro de la misma prioridad, se respeta el orden de llegada.
    def add_client(self, name, priority):
        nuevo = self.Node(name, priority)
        # CASO 1: Insertar al principio (Lista vacía o nuevo VIP ante primer Normal)
        if self.first is None or (priority == 1 and self.first.priority == 2):
            nuevo.next = self.first
            self.first = nuevo
        else:
            # CASO 2: Buscar el hueco en medio o al final
            actual = self.first
            while actual.next is not None:
                # Si el nuevo es VIP y el siguiente es Normal, ¡hemos encontrado el hueco!
                if priority == 1 and actual.next.priority == 2:
                    break
                # Si el nuevo es Normal, seguimos hasta el final (None)
                actual = actual.next
            # Al salir del bucle, 'actual' es el nodo ANTERIOR al hueco
            nuevo.next = actual.next
            actual.next = nuevo
        self._size += 1

    # attend_client: Elimina y devuelve el nombre del primer cliente de la lista (mayor prioridad y más tiempo esperando). Devuelve None si la cola está vacía.
    def attend_client(self):
        if self.first is None:
            return None
        nombre = self.first.name  # Guardamos datos
        self.first = self.first.next  # Siguiente nodo
        self._size -= 1
        return nombre

    # promote_to_vip: Busca un cliente por nombre y, si es de prioridad 2, lo elimina de su posición actual y lo reinserta al final de la sección VIP.
    def promote_to_vip(self, name):
        actual = self.first
        anterior = None
        encontrado = False
        while actual is not None:
            if actual.name == name:
                if actual.priority == 2:
                    if anterior is None:
                        self.first = actual.next
                    else:
                        anterior.next = actual.next
                    self._size -= 1
                    encontrado = True
                break
            anterior = actual
            actual = actual.next
        if encontrado:
            self.add_client(name, 1)

    # vip_count: Devuelve el número de clientes con prioridad 1 en la cola.
    @property
    def vip_count(self):
        count = 0
        actual = self.first
        while actual is not None and actual.priority == 1:
            count += 1
            actual = actual.next
        return count
```

### Gestión de Expedientes Académicos
Se pide completar la clase AcademicRecord que gestiona las asignaturas de un estudiante. Cada asignatura está representada por la clase Subject, que tiene tres atributos: code (int), name (str) y grade (float).La clase debe usar internamente una lista enlazada de objetos Node.
✅ Métodos y propiedades a desarrollar:
- Método add_subject(self, subject)
    - Añade una asignatura al principio de la lista. 
- Método withdraw_subject(self, code)
    - Busca la asignatura por su code y la elimina de la lista. Debe devolver True si se eliminó y False si no se encontró.
- Método move_to_front(self, code)
    - Busca la asignatura con el código indicado. Si la encuentra y no es ya la primera, debe "descolgarla" de su posición actual y reinsertarla al principio de la lista.
    - Pista: Es una combinación de buscar con anterior, saltar el nodo y actualizar self.first.
- Propiedad average_grade
    - Calcula y devuelve la nota media de todas las asignaturas en el expediente. Si no hay asignaturas, devuelve 0.0.

```python
from containerbase import ContainerBase # type: ignore

class Subject:
    def __init__(self, code, name, grade):
        self.code = code
        self.name = name
        self.grade = grade

class AcademicRecord(ContainerBase):
    class Node:
        def __init__(self, value, next=None):
            self.value = value
            self.next = next

    def __init__(self):
        self.first = None # Objeto inicial
        self._size = 0    # Atributo de longitud
    
    # add_subject: Añade una asignatura al principio de la lista.
    def add_subject(self, subject):
        nuevo = self.Node(subject)        
        if self.first is None:
            self.first = nuevo
        else:
            self.first = self.first.next
            self.first = nuevo
    
    # withdraw_subject: Busca una asignatura por su código y la elimina de la lista. Devuelve True si se eliminó y False si no se encontró.
    def withdraw_subject(self, code):
        actual = self.first
        anterior = None
        while actual is not None:
            if actual.value.code == code:
                if anterior is None:
                    self.first = actual.next
                else:
                    anterior.next = actual.next

                self._size -= 1
                return True
            anterior = actual
            actual = actual.next
        return False
    
    # move_to_front: Busca una asignatura por su código y, si no es la primera, la elimina de su posición actual y la reinserta al principio de la lista.
    def move_to_front(self, code):
        if self.first is None or self.first.value.code == code:
            return
        actual = self.first
        anterior = None
        while actual is not None:
            if actual.value.code == code:
                anterior.next = actual.next
                actual.next = self.first
                self.first = actual
                return
            anterior=actual
            actual=actual.next

    # average_grade: Calcula y devuelve la nota media de todas las asignaturas. Si no hay asignaturas, devuelve 0.0.
    @property
    def average_grade(self):
        if self.first is None:
            return 0.0
        suma_notas = 0
        actual = self.first
        while actual is not None:
            suma_notas += actual.value.grade
            actual = actual.next
        return suma_notas / self._size  
```
